```python
import numpy as np
from criteria.multi_criteria import *
```

```python
matrix = np.array([
    [459, 510, 6400,  16.2],
    [645, 570, 12800, 18.0],
    [639, 558, 51200, 20.1],
    [548, 492, 16000, 20.1],
    [677, 375, 25600, 20.3],
    [399, 247, 6400,  16.0]
])

alternatives = [
    "Nikon D5100", 
    "Canon EOS 700D", 
    "Pentax K-S1", 
    "Sony ALPHA SLT A58K", 
    "Samsung NX30", 
    "OLYMPUS TG-3"
]

maximization = [False, False, True, True]
```

```python
print("--- 1. Metóda entropie ---")
vahy_entropia = metoda_entropie(matrix, info=True)
print(f"-> Výsledné váhy (Entropia): {np.round(vahy_entropia, 4)}\n")

print("--- 2. Metóda poradia ---")
vahy_poradia = metoda_poradia(ranks=[3, 4, 1, 2], info=True)
print(f"-> Výsledné váhy (Poradie): {np.round(vahy_poradia, 4)}\n")

print("--- 3. Fullerova metóda ---")
vahy_fuller = fullerova_metoda(frequencies=[3, 2, 1, 0], modified=True, info=True)
print(f"-> Výsledné váhy (Fuller - modifikovaná): {np.round(vahy_fuller, 4)}\n")

print("--- 4. Bodovacia metóda (pre váhy) ---")
vahy_bodovacia = bodovacia_metoda_vahy(body=[80, 50, 90, 60])
print(f"-> Výsledné váhy (Bodovacia): {np.round(vahy_bodovacia, 4)}\n")

print("--- 5. Saatyho metóda (pre váhy kritérií) ---")
S_matica = np.array([[1,   3,   1/3, 2],[1/3, 1,   1/5, 1/2],
    [3,   5,   1,   4],[1/2, 2,   1/4, 1]
])
vahy_saaty = saatyho_metoda_vahy(S_matica, info=True)
print(f"-> Výsledné váhy (Saaty): {np.round(vahy_saaty, 4)}\n")

print("--- 6. Postupný rozvrh váh ---")
vahy_postupne = postupny_rozvrh_vah(
    skupiny_vahy=[0.6, 0.4], 
    kriteria_v_skupinach_vahy=[[0.7, 0.3], [0.8, 0.2]],
    info=True
)
print(f"-> Výsledné váhy (Postupný rozvrh): {np.round(vahy_postupne, 4)}\n")
```

```python
test_weights = vahy_postupne

aspiracne_urovne = [600, 500, 12000, 18.0]

print("--- 7. Konjunktívna metóda ---")
idx_konj = konjunktivna_metoda(matrix, aspiracne_urovne, maximalizacne=maximization)
print(f"-> Akceptované varianty (Konjunktívna): {[alternatives[i] for i in idx_konj]}\n")

print("--- 8. Disjunktívna metóda ---")
idx_disj = disjunktivna_metoda(matrix, aspiracne_urovne, maximalizacne=maximization)
print(f"-> Akceptované varianty (Disjunktívna): {[alternatives[i] for i in idx_disj]}\n")

print("--- 9. Metóda PRIAM ---")
idealny_bod =[399, 247, 51200, 20.3]
idx_priam, msg_priam, _ = metoda_priam_krok(
    matrix, 
    z_nove=aspiracne_urovne, 
    idealny_variant=idealny_bod, 
    maximization=maximization, 
    info=True
)
if isinstance(idx_priam, (list, np.ndarray)):
    print(f"-> Výsledok PRIAM: {[alternatives[i] for i in idx_priam]} ({msg_priam})\n")
else:
    print(f"-> Výsledok PRIAM: {alternatives[idx_priam]} ({msg_priam})\n")
```

```python
print("--- 10. Metóda poradia variantov ---")
idx_poradia, P_mat, weighted_ranks = metoda_poradia_varianty(matrix, test_weights, maximization=maximization)
print(f"Matica poradí variantov:\n{P_mat}")
print(f"Vážené súčty poradí: {np.round(weighted_ranks, 4)}")
print(f"-> Najlepší variant: {alternatives[idx_poradia]}\n")
```

```python
print("--- 11. Lexikografická metóda ---")
# Poradie dôležitosti indexov: ISO(2) -> Cena(0) -> Rozlíšenie(3) -> Hmotnosť(1)
idx_lex = lexikograficka_metoda(matrix, poradie_kriterii=[2, 0, 3, 1], maximization=maximization)
print(f"Zostávajúce indexy po postupnom filtrovaní: {idx_lex}")
print(f"-> Najlepší variant(y): {[alternatives[i] for i in idx_lex]}\n")
```

```python
print("--- 12. Metóda váženého súčtu ---")
idx_wsa, U_mat, scores_wsa = metoda_vazeneho_suctu(matrix, test_weights, maximization=maximization)
print(f"Normalizovaná matica (U):\n{np.round(U_mat, 4)}")
print(f"Výsledné skóre variantov (G_i): {np.round(scores_wsa, 4)}")
print(f"-> Najlepší variant: {alternatives[idx_wsa]}\n")
```

```python
print("--- 13. Bodovacia metóda (pre varianty) ---")
scored_mock = np.random.randint(1, 11, size=matrix.shape)
print(f"Náhodne priradené body pre maticu:\n{scored_mock}")
idx_bod_var, G_i_bod = bodovacia_metoda_varianty(scored_mock, test_weights)
print(f"Vážené súčty bodov pre jednotlivé varianty: {np.round(G_i_bod, 4)}")
print(f"-> Najlepší variant: {alternatives[idx_bod_var]}\n")
```

```python
print("--- 14. Permutačná metóda ---")
best_perm, r_phi, C_mat = permutacna_metoda(matrix, test_weights, maximization=maximization)
print(f"Matica indexov súhlasu (C):\n{np.round(C_mat, 4)}")
print(f"Hodnota R_phi pre najlepšiu permutáciu: {r_phi}")
print(f"-> Najlepšie celkové poradie variantov: {[alternatives[i] for i in best_perm]}\n")
```

```python
print("--- 15. Metóda TOPSIS ---")
idx_topsis, c_i, d_plus, d_minus = metoda_topsis(matrix, test_weights, maximization=maximization)
print(f"Vzdialenosti od ideálneho variantu (d+): {np.round(d_plus, 4)}")
print(f"Vzdialenosti od bazálneho variantu (d-): {np.round(d_minus, 4)}")
print(f"-> Relatívne ukazovatele c_i: {np.round(c_i, 4)}")
print(f"-> Najlepší variant: {alternatives[idx_topsis]}\n")
```

```python
print("--- 16. Metóda ORESTE ---")
# Ako vstupné poradia variantov pre ORESTE použijeme P_mat, ktorú sme získali z Metódy poradia
idx_oreste, r_sums, R_mat, D_mat = metoda_oreste_hodnotenie(q_ranks=[3, 4, 1, 2], P_matrix=P_mat, r=3)
print(f"Matica vzdialeností (D):\n{np.round(D_mat, 4)}")
print(f"Matica globálnych poradí (R):\n{R_mat}")
print(f"Súčty poradí pre varianty: {r_sums}")
print(f"-> Najlepší variant: {alternatives[idx_oreste]}\n")
```

```python
print("--- 17. Metóda ELECTRE I ---")
eff_variants, P_matrix_electre, C_idx, D_idx = metoda_electre_I(matrix, test_weights, c_star=0.6, d_star=0.4, maximization=maximization)
print(f"Matica indexov zhody (C_ij):\n{np.round(C_idx, 4)}")
print(f"Matica indexov nezhody (D_ij):\n{np.round(D_idx, 4)}")
print(f"Výsledná matica preferencií P (0/1):\n{P_matrix_electre}")
if eff_variants:
    print(f"-> Efektívne (nedominované) varianty: {[alternatives[i] for i in eff_variants]}\n")
else:
    print("-> Nenašli sa jednoznačne efektívne varianty pre dané prahy c* a d*.\n")
```

```python
print("--- 18. Metóda AGREPREF ---")
agrepref_relations = metoda_agrepref(matrix, test_weights, alpha=0.5, beta=0.1, maximization=maximization)
print(f"Výsledná matica relácií AGREPREF (1=P, 2=I, 3=N):\n{agrepref_relations}")
```

```python
print("--- 19. Metóda AHP (pre varianty) ---")
saaty_krit_mock = np.ones((4, 4)) 
zoznam_saaty_var_mock =[np.ones((6, 6)) for _ in range(4)]
idx_ahp_var, ahp_scores, var_weights_mat = metoda_ahp_varianty(saaty_krit_mock, zoznam_saaty_var_mock)
print(f"Lokálne váhy variantov pre každé kritérium:\n{np.round(var_weights_mat, 4)}")
print(f"Celkové skóre variantov:\n{np.round(ahp_scores, 4)}")
print(f"-> Najlepší variant: {alternatives[idx_ahp_var]}\n")
```
